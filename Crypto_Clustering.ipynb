```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas    
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

df_market_normalized = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data

df_market_scaled = pd.DataFrame(
    df_market_normalized,
columns=df_market_data.columns
# columns=["count", "mean", "std", "min", "25%", "50%","75%", "max"]
    )

# Copy the crypto names from the original data
df_market_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_scaled = df_market_scaled.set_index("coin_id")

# Display sample data
df_market_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
# model = KMeans(n_clusters=3, random_state=1)
k_values  = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for n in k_values:
    model= KMeans(n_clusters=n, random_state=0)
    model.fit(df_market_scaled)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve = {"k_values" :k_values, "inertia":inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_curve_df = pd.DataFrame(elbow_curve)

elbow_curve_df.head(10)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_curve_df.hvplot.line(
    x="k_values", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k_values
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = model.predict(df_market_scaled)


# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame
df_market_scaled_predictions = df_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters

df_market_scaled_predictions["CryptoCluster"] = crypto_clusters


# Display sample data
df_market_scaled_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_scaled_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="CryptoCluster",
    hover_cols = ["coin_id"], 
    title = "Scatter Plot by Crypto Segment - k=4"
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components = 3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca_data = pca.fit_transform(df_market_scaled_predictions)
# View the first five rows of the DataFrame. 
crypto_pca_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
pca.explained_variance_ratio_.sum()
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Total explained variance is 88.68%

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(crypto_pca_data, columns=['PC1', 'PC2', 'PC3'])

# Creating a DataFrame with the PCA data

# Copy the crypto names from the original data
df_market_data_pca["coin_id"] = df_market_scaled.index

# Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index("coin_id")

# Display sample data
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k  = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_data_pca)
    inertia_pca.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    "k": k,
    "inertia_pca": inertia_pca
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot_pca = df_elbow_pca.hvplot.line(x="k", y="inertia_pca", title="Elbow Curve Using Standardarized Crypto PCA Data", xticks=k)
elbow_plot_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Best value is 4 when using the PCA data


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No change in the best value. It remains 4

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_cluster = model_pca.predict(df_market_data_pca)
# Print the resulting array of cluster values.
print(pca_cluster)
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_data_pca_prediction = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_pca_prediction["PCACluster"] = pca_cluster

# Display sample data
df_market_data_pca_prediction.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_data_pca_prediction.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="PCACluster",
    title = "Scatter Plot by Stock Segment - PCA=2"
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Create a composite plot to visualize the Elbow curves for both the original scaled DataFrame and the PCA scaled DataFrame
# Elbow Curve for Original Scaled DataFrame
elbow_original = elbow_curve_df.hvplot.line(
    x='k_values', 
    y='inertia', 
    label='Original Scaled Data', 
    color='blue'
)

# Elbow Curve for PCA Scaled DataFrame
elbow_pca = df_elbow_pca.hvplot.line(
    x='k', 
    y='inertia_pca', 
    label='PCA Scaled Data', 
    color='orange'
)

# Stack plots vertically using the * operator
stacked_plot = elbow_original * elbow_pca

# Display the stacked plot
stacked_plot
```

```python
# Create a composite plot to visualize the Elbow curves for both the original scaled DataFrame and the PCA scaled DataFrame
# Scatter plot for original scaled clusters
clusters_original = df_market_scaled_predictions.hvplot.scatter(
    x='price_change_percentage_24h', 
    y='price_change_percentage_7d', 
    by='CryptoCluster', 
    colormap='viridis', 
    title='Original Scaled Data'
)

# Scatter plot for PCA scaled clusters
clusters_pca = df_market_data_pca_prediction.hvplot.scatter(
    x='PC1', 
    y='PC2', 
    by='PCACluster', 
    colormap='plasma', 
    title='PCA Scaled Data'
)

# Combine both plots using the + operator
composite_plot = clusters_original + clusters_pca

# Display the composite plot
composite_plot
```
